```python
def crop_image(im, new_width, new_height):
    width, height = im.size   # Get dimensions

    left = (width - new_width)/2
    top = (height - new_height)/2
    right = (width + new_width)/2
    bottom = (height + new_height)/2

    # Crop the center of the image
    im = im.crop((left, top, right, bottom))
    return im

def crop_arr(arr, new_width, new_height):
    width, height = arr.shape[1], arr.shape[0]   # Get dimensions

    left = int((width - new_width)/2)
    top = int((height - new_height)/2)
    right = int((width + new_width)/2)
    bottom = int((height + new_height)/2)

    # Crop the center of the array
    cropped = arr[top:bottom, left:right, :]
    return cropped

def save_image(im, path):
    """
        Saves a numpy matrix or PIL image as an image
    Args:
        im_as_arr (Numpy array): Matrix of shape DxWxH
        path (str): Path to the image
    """
    if isinstance(im, (np.ndarray, np.generic)):
        # im = format_np_output(im)
        im = Image.fromarray(im)
    im.save(path)
```

```python
class Fig:
    def __init__(self, crop):
        spectra = 0
        value = np.zeros((224, 224))
        self.value = value
        if crop == 1:
            self.colors = {'orange' : spectra, 'lblue' : spectra, 'blue' : spectra, 'green' : spectra}
        else:
            self.colors = {'pink' : spectra, 'purple' : spectra, 'red' : spectra, 'yellow' : spectra}

class Illuminant:
    def __init__(self, illuminant):
        self.images = {illuminant : Fig(1), illuminant + '2' : Fig(2), 'm1': Fig(1), 'm2': Fig(2)}

def get_errors(illuminant, method, error_type):
    I = Illuminant(illuminant)
    print(I.images.keys())
    for image in I.images.keys():
        I.images[image].value = np.array(Image.open('./' + error_type + '_maps/' + method + '/' + image + '.png'))
        values = I.images[image].value.copy() / 255
        if image == illuminant or image == 'm1':
            I.images[image].colors['orange'] = np.mean(values[75:125, 35:60])
            I.images[image].colors['lblue'] = np.mean(values[75:125, 140:160])
            I.images[image].colors['blue'] = np.mean(values[175:200, 0:50])
            I.images[image].colors['green'] = np.mean(values[175:200, 100:150])
        else:
            I.images[image].colors['pink'] = np.mean(values[90:125, 35:60])
            I.images[image].colors['purple'] = np.mean(values[90:125, 140:160])
            I.images[image].colors['red'] = np.mean(values[175:200, 0:50])
            I.images[image].colors['yellow'] = np.mean(values[175:200, 100:150])
    return I

```
